```python
import math

y_true = [1,0,1,1,0]
y_predicted = [0.9,0.8,0.7,0.3,0.2]
```

```python
epsilon = 1e-15
log_losses = []

for yt,yp in zip(y_true, y_predicted):
    yp = min(max(yp,epsilon),1-epsilon)
    loss = yt*math.log(yp)+(1-yt)*math.log(1-yp)
    log_losses.append(loss)

log_loss = -sum(log_losses)/len(y_true)
print("log loss",log_loss)
```

```python
import numpy as np
y_true = np.array([1,0,1,1,0])
y_predict = np.array([0.9,0.8,0.7,0.3,0.2])

epsilon = 1e-15
y_predict = np.clip(y_predict,epsilon,1-epsilon)

log_loss = -np.mean(y_true*np.log(y_predict) + (1-y_true)*np.log(1-y_predict))

print("log loss", log_loss)
```
